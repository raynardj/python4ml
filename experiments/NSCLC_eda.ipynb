```python
patients.vc("acronym")
```

Filter out the NSCLC types

```python
nsclc = patients[patients.acronym.isin(["LUAD","LUSC"])].reset_index(drop=True)
```

```python
nsclc
```

```python
from ipywidgets import interact
```

```python
@interact
def show_distribution(kw = "search"):
    
    matchlist = [col for col in nsclc.columns if kw in col]
    @interact_manual
    def show_dis(col_ = matchlist):
        return nsclc.vc(col_)
```

```python
from plotly import express as px
```

```python
vc_label = nsclc.vc("tobacco_smoking_history")
px.bar(vc_label, y="tobacco_smoking_history", x=list(vc_label.index))
```
